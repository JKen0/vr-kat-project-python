```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the number of points and the range
num_points = 20
x = np.linspace(0, 2 * np.pi, num_points)

# Calculate the corresponding y values using the sine function
y = np.sin(x)
```

```python
# Plot the discrete points
plt.plot(y, marker='o', linestyle='-', color='b')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Discrete Sine Wave (1 Cycle)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Rearrange the y values
rearranged_y = np.concatenate((y[::2], y[1::2]))

# Plot the discrete points
plt.plot(y, marker='o', linestyle='-', color='b', label='Original')
plt.plot(rearranged_y, marker='o', linestyle='-', color='g', label="Rearranged")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Rearranged Sine Wave (Double Speed)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Interpolate to double the cycle length
interpolated_y = np.empty(2 * num_points - 1)
for i in range(num_points - 1):
    interpolated_y[2 * i] = y[i]
    interpolated_y[2 * i + 1] = (y[i] + y[i + 1]) / 2.0
interpolated_y[-1] = y[-1]

interpolated_x = np.linspace(0, 4 * np.pi, 2*num_points - 1)

# Plot the interpolated points
plt.plot(y, marker='o', linestyle='-', color='b', label='Original')
plt.plot(interpolated_y, marker='o', linestyle='-', color='g', label='Interpolated')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Interpolated Sine Wave (Half-Speed)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Define the x values from 0 to 20π to cover 10 cycles
x = np.linspace(0, 20 * np.pi, 1000)

# Calculate the corresponding y values for the sine wave
y = np.sin(x)

# Create a horizontal line at y = 0.9
threshold = 0.9
horizontal_line = np.full_like(x, threshold)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Sine Wave')
plt.plot(x, horizontal_line, linestyle='--', color='red', label='Threshold')
plt.xlabel('x')
plt.ylabel('y')
plt.title('10 Cycles of Sine Wave with Threshold')
plt.legend()
plt.grid()
plt.show()
```
