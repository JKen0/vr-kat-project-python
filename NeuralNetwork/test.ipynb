```python
import os
import pandas as pd
import numpy as np
from scipy.signal import find_peaks
import matplotlib.pyplot as plt
```

```python
file_name = 'C:\Dev\\vr-kat-project-python-2\processed-training-data\\4-PROCESSED-DATA\PROCESSED-LEFT-STEPS.xlsx'


df = pd.read_excel(file_name)  # Load each Excel file

df = df.head(1000)

# Get the sensor values from the DataFrame
sensor_values = df['L_Roll']

# Apply peak detection to identify peaks
peaks, _ = find_peaks(sensor_values, distance=50, prominence=0.1)

# Plot the sensor data with detected peaks
plt.plot(sensor_values)
plt.plot(peaks, sensor_values[peaks], 'ro')
plt.xlabel('Iteration')
plt.ylabel('Sensor Value')
plt.title('Sensor Data with Detected Peaks')
plt.show()
```
