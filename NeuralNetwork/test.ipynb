```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_name = 'C:\Dev\\vr-kat-project-python-2\processed-training-data\\4-PROCESSED-DATA\\TRAIN2\\PROC-TRAIN2-STEPS-LR-LAR-62BPM.xlsx'
file_name2 = 'C:\Dev\\vr-kat-project-python-2\processed-training-data\\4-PROCESSED-DATA\\TRAIN2\\PROC-TRAIN2-STEPS-LR-LAR-60BPM.xlsx'
file_name3 = 'C:\Dev\\vr-kat-project-python-2\processed-training-data\\4-PROCESSED-DATA\\TRAIN2\\PROC-TRAIN2-STEPS-LR-SML-60BPM.xlsx'

ITERATION_SIZE = 700

df = pd.read_excel(file_name)  # Load each Excel file
df2 = pd.read_excel(file_name2)
df3 = pd.read_excel(file_name3)

df = df.head(ITERATION_SIZE)
df2 = df2.head(ITERATION_SIZE)
df3 = df3.head(ITERATION_SIZE)

# Get the sensor values from the DataFrame
sensor_values = df['L_Pitch']
sensor_values2 = df2['L_Pitch']
sensor_values3 = df3['L_Pitch']

# Plot the sensor data with detected peaks
plt.plot(sensor_values, label='large')
plt.plot(sensor_values2, label='medium')
plt.plot(sensor_values3, label='small')
#plt.plot(peaks, sensor_values[peaks], 'ro')
plt.xlabel('Iteration')
plt.ylabel('Sensor Value')
plt.title('Sensor Data and Cycles')
plt.legend()
plt.show()
```
