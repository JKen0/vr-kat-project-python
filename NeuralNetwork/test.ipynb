```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
figure_size_width = 10
figure_size_height = 8
```

```python
file_name = PARENT_DIRECTORY + '\processed-training-data\\4-PROCESSED-DATA\\TRAIN2\\PROC-TRAIN2-STEPS-LR-LAR-62BPM.xlsx'
file_name2 = PARENT_DIRECTORY + '\processed-training-data\\4-PROCESSED-DATA\\TRAIN2\\PROC-TRAIN2-STEPS-LR-SML-60BPM.xlsx'
file_name3 = PARENT_DIRECTORY + '\processed-training-data\\4-PROCESSED-DATA\\EXTRA\\PROC-EXTRA-BSTEPS-LR-LAR-60BPM.xlsx'

ITERATION_SIZE = 700

df = pd.read_excel(file_name)  # Load each Excel file
df2 = pd.read_excel(file_name2)
df3 = pd.read_excel(file_name3)

df = df.head(ITERATION_SIZE)
df2 = df2.head(ITERATION_SIZE)
df3 = df3.head(ITERATION_SIZE)

# Get the sensor values from the DataFrame
sensor_values1 = df[['L_Pitch', 'L_Roll', 'R_Pitch', 'R_Roll']]
sensor_values2 = df2[['L_Pitch', 'L_Roll', 'R_Pitch', 'R_Roll']]
sensor_values3 = df3[['L_Pitch', 'L_Roll', 'R_Pitch', 'R_Roll']]

# Create a horizontal line at y = 0.9
x_values = np.arange(0, ITERATION_SIZE)
constant_y = np.full_like(x_values, 37)

# Create a 2x2 subplot
plt.figure(figsize=(10, 8))

# Loop through each sensor and plot its data from sensor_values1 and sensor_values2
sensors = ['L_Pitch', 'R_Pitch']
for i, sensor in enumerate(sensors, 1):
    plt.subplot(2, 2, i)
    plt.plot(sensor_values1[sensor], label='Large', alpha=0.6)
    plt.plot(sensor_values2[sensor], label='Small', alpha=0.6)
    #plt.plot(sensor_values3[sensor], label='BSTEPS', alpha=0.6)
    plt.plot(x_values, constant_y, linestyle='--', color='red', label='Threshold')
    plt.xlabel('Iteration')
    plt.ylabel('Rotation Value (Sensor Value)')
    plt.title(f'{sensor} Sensor Data')
    plt.legend()


plt.suptitle('Large vs. Small Steps Sensor Data')

# Adjust layout to avoid overlapping titles
plt.tight_layout()

# Show the plot
plt.show()
```

```python
file_name = PARENT_DIRECTORY + '\\NeuralNetwork\\np-saved-data\\training\\ALL-X-TRAIN-CLASSIFY-MOTIONSPEED-LAR-STEPS-V0.npy'
file_name2 = PARENT_DIRECTORY + '\\NeuralNetwork\\np-saved-data\\training\\ALL-Y-TRAIN-CLASSIFY-MOTIONSPEED-LAR-STEPS-V0.npy'

ALL_X_DATA = np.load(file_name)
ALL_Y_DATA = np.load(file_name2)

SUM_X_DATA = np.sum(ALL_X_DATA[:, [0, 2]], axis=1)

print(SUM_X_DATA.shape)
print(ALL_Y_DATA.shape)

# Example data: Replace these with your actual input and class arrays
input_data = np.copy(SUM_X_DATA)
class_labels = np.copy(ALL_Y_DATA)

# Dictionary to store data for each class
class_data = {}
for label in np.unique(class_labels):
    class_data[label] = input_data[class_labels == label]

# Box plot
plt.figure(figsize=(5, 4))
plt.boxplot([class_data[label] for label in np.unique(class_labels)],
            labels=np.unique(class_labels),
            patch_artist=True,
            boxprops=dict(facecolor='lightblue'),
            medianprops=dict(color='black'))
plt.xlabel("")
plt.ylabel("Input Data")
plt.title("Box Plot of Input Data for Different Classes")
plt.grid(True)
plt.show()
```
